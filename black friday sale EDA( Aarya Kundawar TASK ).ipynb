```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv("BlackFriday.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df['Total_Product_Categories'] = df[['Product_Category_1', 'Product_Category_2', 'Product_Category_3']].count(axis=1)
```

```python
scaler = MinMaxScaler()
df['Purchase_Scaled'] = scaler.fit_transform(df[['Purchase']])
```

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Gender', 'Age', 'City_Category']]).toarray()
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Gender', 'Age', 'City_Category']))
df = pd.concat([df, encoded_df], axis=1)
```

```python
print(df.head())
```

```python
#correlation analysis
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
correlation_matrix = df[numerical_cols].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Hypothesis
```

```python
from scipy.stats import f_oneway, ttest_ind
```

```python
anova_result = f_oneway(df['Purchase'][df['Age'] == '0-17'],
                         df['Purchase'][df['Age'] == '18-25'],
                         df['Purchase'][df['Age'] == '26-35'],
                         df['Purchase'][df['Age'] == '36-45'],
                         df['Purchase'][df['Age'] == '46-50'],
                         df['Purchase'][df['Age'] == '51-55'],
                         df['Purchase'][df['Age'] == '55+'])
```

```python
ttest_result = ttest_ind(df['Purchase'][df['Marital_Status'] == 0],
                          df['Purchase'][df['Marital_Status'] == 1])
```

```python
print("Hypothesis 1 (Difference in mean purchase amount between age groups):")
print("ANOVA p-value:", anova_result.pvalue)
print("\nHypothesis 2 (Effect of marital status on purchase behavior):")
print("t-test p-value:", ttest_result.pvalue)
```

```python
from scipy.stats import f_oneway, ttest_ind

df.dropna(subset=['Purchase', 'Age', 'Marital_Status'], inplace=True)

anova_result = f_oneway(df['Purchase'][df['Age'] == '0-17'],
                         df['Purchase'][df['Age'] == '18-25'],
                         df['Purchase'][df['Age'] == '26-35'],
                         df['Purchase'][df['Age'] == '36-45'],
                         df['Purchase'][df['Age'] == '46-50'],
                         df['Purchase'][df['Age'] == '51-55'],
                         df['Purchase'][df['Age'] == '55+'])

ttest_result = ttest_ind(df['Purchase'][df['Marital_Status'] == 0],
                          df['Purchase'][df['Marital_Status'] == 1])

print("Hypothesis 1 (Difference in mean purchase amount between age groups):")
print("ANOVA p-value:", anova_result.pvalue)

print("\nHypothesis 2 (Effect of marital status on purchase behavior):")
print("t-test p-value:", ttest_result.pvalue)
```
